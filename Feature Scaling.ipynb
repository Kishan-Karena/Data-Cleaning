```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("loan_sanction_test.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
sns.boxplot(x="ApplicantIncome",data=data)
plt.show()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(data[["ApplicantIncome"]])
```

```python
data["ApplicantIncome_new"]=pd.DataFrame(ss.transform(data[["ApplicantIncome"]]),columns=["x"])
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.boxplot(x="ApplicantIncome_new",data=data)
plt.subplot(1,2,2)
sns.boxplot(x="ApplicantIncome",data=data)
plt.show()
```

```python
# Using min , max and min
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(data[["CoapplicantIncome"]])
```

```python
data["CoapplicantIncome_min"]=ms.transform(data[["CoapplicantIncome"]])
```

```python
data.head()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.boxplot(x="CoapplicantIncome_min",data=data)
plt.subplot(1,2,2)
sns.boxplot(x="CoapplicantIncome",data=data)
plt.show()
```
